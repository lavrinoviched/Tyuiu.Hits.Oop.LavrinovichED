```polyglot-notebook
using System;

public class Plant
{
    public string Name { get; set; }
    public string Type { get; set; }
    public float Height { get; set; }

    public Plant(string name, string type, float height)
    {
        Name = name;
        Type = type;
        Height = height;
    }

    public virtual void Bloom()
    {
        Console.WriteLine($"Растение {Name} цветет");
    }

    public virtual void Grow(float growth)
    {
        if (growth > 0)
        {
            Height += growth;
            Console.WriteLine($"{Name} выросло на {growth} см. Высота составляет: {Height} см");
        }
        else
        {
            Console.WriteLine("Рост должен быть положительным числом, введите данные ещё раз ");
        }
    }
}
public class Flower : Plant
{
    public string Color { get; set; }
    public int PetalsCount { get; set; }

    public Flower(string name, string color, int petalsCount, float height)
        : base(name,"Цветок", height)
    {
        Color = color;
        PetalsCount = petalsCount;
    }

    public override void Bloom()
    {
        Console.WriteLine($"{Name} распускается!");
        Console.WriteLine($"Цвет: {Color}, Лепестков: {PetalsCount}");
    }
    
    public void Wither()
    {
        Console.WriteLine($"{Name} начинает увядать");
    }
}

public class Tree : Plant
{

    public int Age { get; set; }
    public string LeavesType { get; set; }

    public Tree(string name, int age, string leavesType, float height)
        : base(name, "Дерево", height)
    {
        Age = age;
        LeavesType = leavesType;
    }

    public override void Bloom()
    {
        Console.WriteLine($"{Name} покрывается новыми листьями!");
        Console.WriteLine($"Тип листьев: {LeavesType}");
    }

    public void GrowAge(int years)
    {
        if (years > 0)
        {
            Age += years;
            Console.WriteLine($"{Name} стало старше на {years} лет. Возраст составляет: {Age} лет");
        }
        else
        {
            Console.WriteLine("Количество лет должно быть положительным, введите данные ещё раз");
        }
    }
}
Flower pion = new Flower("Пион", "Розовый", 12, 45);
Tree oak = new Tree("Ель", 15, "Хвойные", 850);

Console.WriteLine("Цветок");
pion.Bloom();           
pion.Grow(10);      
pion.Wither();        
Console.WriteLine();
Console.WriteLine("Дерево");
oak.Bloom();            
oak.Grow(25);        
oak.GrowAge(2);        
Console.WriteLine();







```
